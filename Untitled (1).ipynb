```python
# make a taylor expanded cos function
def tayl_cos(x):
    return 1-1/2*x**2+1/(4*3*2)*x**4

#generate a set of x values
t_arr = linspace(0,3,3001)

#make the y values of the taylor expanded and normal cos function using x values
y1 = tayl_cos(t_arr)
y2 = cos(t_arr)

#plot them over each other
plot(t_arr,y1)
plot(t_arr,y2)

#make legend
legend(["taylor","exact"])

#set the index to be the last element in the array
index = len(y1)
for i in range(len(y1)):
    #find the first index where the error is greater than 5%
    if abs(y1[i] - y2[i])/y2[i] > 0.05 and i < index:
        index = i
        
#print index times the spacing of each x value to get the distance
print(index * 0.001)
```

```python
#define taylor expanded exponential
def tayl_exp(x):
    return 1+x+x**2/2+x**3/(6)

#generate new larger x linspace
t_arr = linspace(0,2,2001)

#generate the y values for the taylor and normal exponential
y1 = tayl_exp(t_arr)
y2 = exp(t_arr)

#plot
plot(t_arr,y1)
plot(t_arr,y2)
legend(["taylor","exact"])

#get the index again
index = len(y1)
for i in range(len(y1)):
    if abs(y1[i] - y2[i])/y2[i] > 0.05 and i < index:
        index = i

#multiply and print to get the x distance
print(index * 0.001)
```

```python
#make a factorial function
def fact(n):
    ans = 1
    for i in range(1,n+1):
        ans = ans*i
    return ans

#make a more genereal taylor exponential to save time
#it takes the order and what its centered around
def tayl_exp(x,x_0,n):
    ans = 0
    for i in range(n+1):
        ans += exp(x_0)*(x-x_0)**i/fact(i)
    return ans


#make new x linspace
t_arr = linspace(0,5,5001)

#generate a ninth order approximation
y1 = tayl_exp(t_arr,0,9)
y2 = exp(t_arr)

#plot
plot(t_arr,y1)
plot(t_arr,y2)
legend(["taylor","exact"])

#check if there is a 5% error anywhere
index = len(y1)
for i in range(len(y1)):
    if abs(y1[i] - y2[i])/y2[i] > 0.05 and i < index:
        index = i

#distance is bigger than 5 (This is just what we set it too initially it may not deviate by 5% yet)
print(index * 0.001)
print("you need an order of nine in order to have an error less than 5% at x=5 when centered around x = 0")
```
